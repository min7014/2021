01. Geogebra를 실행시킨다.

02. Input에

    **Dice=1**

    입력하여 Dice=1 이라는 Number를 만든다.

03. Input에

    **DiceThrow=Button()**

    입력하여 DiceThrow Button을 만든다.

04. DiceThrow Button에서 Right Mouse Click --> Object Properties --> Scripting --> On Click

    1번째 줄 : **SetValue[Dice,RandomBetween[1,6]]**

    을 Scripting 창에 입력 --> Object Properties 창 닫기

05. Move Tool를 선택 --> DiceThrow Button 선택

    *Move Tool는 Geogebra 의 Toolbar에 화살표 표시입니다.

06. DiceThrow Button을 선택할 때 마다 Dice 값이 자동으로 변하는 것을 볼 수 있다.

    SetValue과 RandomBetween 이라는 Geogebra Scripting 명령의 의미를 설명해 준다. 경우에 따라서는 Dice의 수가 기본값이 최댓값이 5인 Slider로 되어 6이 표현 안되는 경우가 있다. 이러한 경우 최댓값을 6으로 조정해주면 된다.

07. Input에

   **CollectionOfDiceResults={}**

   입력하여 CollectionOfDiceResults 이라는 빈 list를 만든다.

08. DiceThrow Button에서 Right Mouse Click --> Cbject Properties --> Scripting --> On Click

    1번째 줄 끝을 선택하고 [Enter]를 입력하면 2번째 줄이 생긴다.

    2번째 줄 : **SetValue[CollectionOfDiceResults,Append[CollectionOfDiceResults,Dice]]**

    을 Scripting 에 입력  --> Object Properties 닫기

09. Move Tool를 선택 --> DiceThrow Button 선택

    *Move Tool는 Geogebra 의 Toolbar에 화살표 표시입니다. 

10. DiceThrow Button을 선택할 때 마다 Dice 값이 자동으로 변하는 것을 볼 수 있다. 또한, CollectionOfDiceResults list에 차곡차곡 Dice에서 나온 값들이 쌓이는 것을 볼 수 있다.

11. Input에

    **NumberOfDice1=CountIf[x==1,CollectionOfDiceResults]**

    **NumberOfDice2=CountIf[x==2,CollectionOfDiceResults]**

    **NumberOfDice3=CountIf[x==3,CollectionOfDiceResults]**

    **NumberOfDice4=CountIf[x==4,CollectionOfDiceResults]**

    **NumberOfDice5=CountIf[x==5,CollectionOfDiceResults]**

    **NumberOfDice6=CountIf[x==6,CollectionOfDiceResults]**

    **NumberOfDice=CountIf[x==x,CollectionOfDiceResults]**

    차례로 입력하여 CollectionOfDiceResults에 모아둔 Dice눈의 개수를 자동으로 셀 수 있게 한다.

12. 쓰인 명령어 CountIf에 대하여 이야기 해준다.

    조건으로 쓴 x==1, x==2, ..., x==6에 대하여 알려주고 x==x가 전체 개수를 쓰는 조건임을 알려준다.

13. NumberOfDice가 6번, 60번, 120번, 1200번 정도 했을 때 나온 눈의 개수들이 NumberOfDice의 6분의 1에 가까운 수가 나오는 것을 확인하게 한다.

14. Input에   

    **CollectionOfDiceResultsInitialization=Button()**

    입력하여 CollectionOfDiceResultsInitialization Button을 만든다.

15. Input에

    **SetCoords(CollectionOfDiceResultsInitialization,10,10)**

    **SetCoords(DiceThrow,10,40)**

    을 입력하여 DiceThrow Button과, CollectionOfDiceResultsInitialization Button을 화면에 겹치치 않게 배치한다.

16. CollectionOfDiceResultsInitialization Button에서 Right Mouse Click --> Cbject Properties --> Scripting --> On Click

    1번째 줄 : **SetValue[CollectionOfDiceResults ,{}]**

    을 Scripting 창에 입력 --> Object Properties 창 닫기

17. CollectionOfDiceResultsinitialization Button을 선택하면 CollectionOfDiceResults list가 {}이 되어 모아두었던 Dice의 눈이 사라진다.

01. Geogebra를 실행시킨다.

02. 입력창에

    **주사위=1**

    입력하여 주사위=1 이라는 수를 만든다.

03. 입력창에

    **주사위던지기=버튼()**

    입력하여 주사위던지기 버튼을 만든다.

04. 주사위던지기 버튼에서 오른쪽 마우스 클릭 --> 설정사항 --> 스크립트 --> 클릭할 때

    1번째 줄 : **값설정[주사위,랜덤[1,6]]**

    을 스크립트 창에 입력 --> 확인 --> 설정사항 창 닫기

05. 이동도구를 선택 --> 주사위던지기 버튼 선택

    *이동도구는 Geogebra 상단 도구들 중 화살표 표시입니다.

06. 주사위던지기 버튼을 선택할 때 마다 주사위 값이 자동으로 변하는 것을 볼 수 있다.

    값설정과 랜덤 이라는 Geogebra 스크립트 명령의 의미를 설명해 준다. 경우에 따라서는 주사위가 수가 기본값이 최댓값이 5인 슬라이더로 되어 6이 표현 안되는 경우가 있다. 이러한 경우 최댓값을 6으로 조정해주면 된다.

07. 입력창에

   **주사위결과모음={}**

   입력하여 주사위결과모음 이라는 빈 리스트를 만든다.

08. 주사위던지기 버튼에서 오른쪽 마우스 클릭 --> 설정사항 --> 스크립트 --> 클릭할 때

    1번째 줄 끝을 선택하고 [Enter]를 입력하면 2번째 줄이 생긴다.

    2번째 줄 : **값설정[주사위결과모음,추가[주사위결과모음,주사위]]**

    을 스크립트 창에 입력 --> 확인 --> 설정사항 창 닫기

09. 이동도구를 선택 --> 주사위던지기 버튼 선택

    *이동도구는 Geogebra 상단 도구들 중 화살표 표시입니다. 

10. 주사위던지기 버튼을 선택할 때 마다 주사위 값이 자동으로 변하는 것을 볼 수 있다. 또한, 주사위결과모음 리스트에 차곡차곡 주사위에서 나온 값들이 쌓이는 것을 볼 수 있다.

11. 입력창에

    **주사위1의개수=세기조건[x==1,주사위결과모음]**

    **주사위2의개수=세기조건[x==2,주사위결과모음]**

    **주사위3의개수=세기조건[x==3,주사위결과모음]**

    **주사위4의개수=세기조건[x==4,주사위결과모음]**

    **주사위5의개수=세기조건[x==5,주사위결과모음]**

    **주사위6의개수=세기조건[x==6,주사위결과모음]**

    **주사위전체개수=세기조건[x==x,주사위결과모음]**

    차례로 입력하여 주사위눈의 개수를 자동으로 셀 수 있게 한다.

12. 쓰인 명령어 세기조건에 대하여 이야기 해준다.

    조건으로 쓴 x==1, x==2, ..., x==6에 대하여 알려주고 x==x가 전체개수를 쓰는 조건임을 알려준다.

13. 전체개수가 6번, 60번, 120번, 1200번 정도 했을 때 나온 눈의 개수들이 전체개수의 6분의 1에 해당하는 수가 나오는 것을 확인하게 한다.

14. 입력창에   

    **주사위결과모음초기화=버튼()**

    입력하여 주사위던지기 버튼을 만든다.

15. 입력창에

    **좌표설정(주사위결과모음초기화,10,10)**

    **좌표설정(주사위던지기,10,40)**

    을 입력하여 주사위던기기 버튼과, 주사위결과모음초기화 버튼을 화면에 겹치치 않게 배치한다.

16. 주사위결과모음초기화 버튼에서 오른쪽 마우스 클릭 --> 설정사항 --> 스크립트 --> 클릭할 때

    1번째 줄 : **값설정[주사위결과모음,{}]**

    을 스크립트 창에 입력 --> 확인 --> 설정사항 창 닫기

17. 주사위결과모음초기화 버튼을 선택하면 주사위결과모음 리스트가 {}이 되어 나왔던 주사위의 눈이 사라진다.
