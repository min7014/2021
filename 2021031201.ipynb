    **A=(1,2)**

04. Input에

    **B=(-2,-3)**

    입력하여 B 라는 점을 만든다.

　

05. Input에

    **c=Segment[A,B]**

    입력하여 c 라는 선분을 만든다.

　

06. Input에

    **l:x=x(A)**

    입력하여 l 이라는 y 축에 평행하면서 점 A 을 지나는 직선을 만든다.

    x(A)의 뜻은 점 A 의 x좌표를 뜻한다.

　

07. Input에

    **m:y=y(B)**

    입력하여 m 이라는 x 축에 평행하면서 점 B을 지나는 직선을 만든다.

    x(B)의 뜻은 점 B 의 y좌표를 뜻한다.

　

08. Input에

    **C=Intersect[l,m]**

    입력하여 C 라는 l 과 m 의 교점을 만든다.

　

09. Input에

    **SetVisibleInView[l,1,false]**

    입력하여 l 직선이 보이지 않게 한다.

　

10. Input에

    **SetVisibleInView[m,1,false]**

    입력하여 m 직선이 보이지 않게 한다.

　

11. Input에

    **a=Segment[B,C]**

    입력하여 a 라는 선분을 만든다.

　

12. Input에

    **b=Segment[C,A]**

    입력하여 b 라는 선분을 만든다.

　

13. Input에

    **BCLength=abs[x(A)-x(B)]**

    입력하여 BCLength 라는 수를 만든다.

    abs[x(A)-x(B)]는 x(A)-x(B)의 절대값을 의미한다.

    x(A)-x(B)는 점 A의 x 좌표에서 점 B의 x 좌표를 뺀 것을 의미한다.

　

14. Input에

    **CALength=abs[y(A)-y(B)]**

    입력하여 CALength이 라는 수를 만든다.

    abs[y(A)-y(B)]는 y(A)-y(B)의 절대값을 의미한다.

    y(A)-y(B)는 점 A의 y 좌표에서 점 B의 y 좌표를 뺀 것을 의미한다.

　

　

　

15. Input에

    **AngleC=Angle[A,C,B]**

    입력하여 AngleC 라는 각을 만든다.

    각 C가 직각이므로 삼각형 ABC는 직각 삼각형이다.

　

16. Input에

    **ABLength=sqrt[CALength^2+BCLength^2]**

    입력하여 ABLength 라는 수를 만든다.

　

17. Geogebra 상단에 File, Edit, View, Options, Window, Help에서

    **Options --> Rounding --> 15 Decimal Places**

    ABLength는 소수점 15자리 이상 필요하다. 아마도 무리수일 것이다.

   

18. Geogebra 상단에 File, Edit, View, Options, Window, Help에

    **View --> CAS**
　

19. Cas 창에 1이라고 써있는 수옆 빈칸에

    **sqrt[CALength^2+BCLength^2]**

    입력하고 [Enter]키를 누른다Length
