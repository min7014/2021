    을 입력하여 Dice의 1의 눈의 개수를 자동으로 셀 수 있게 한다.

    **NumberOfDice2=CountIf[x==2,CollectionOfDiceResults]**

    **NumberOfDice3=CountIf[x==3,CollectionOfDiceResults]**

13. 마찬가지로 Input에

    **NumberOfDice=CountIf[x==x,CollectionOfDiceResults]**

    차례로 입력하여 Dice의 전체 개수를 자동으로 셀 수 있게 한다.

    조건으로 쓴 x==1, x==2, ..., x==6에 대하여 알려주고 x==x가 전체개수를 쓰는 조건임을 알려준다.

    **RelativeFrequencyOfDice2=NumberOfDice2/NumberOfDice**

    **RelativeFrequencyOfDice5=NumberOfDice5/NumberOfDice**

    **RelativeFrequencyOfDice6=NumberOfDice6/NumberOfDice**

21. Dice의 던기지 버튼을 눌러 Relative Frequency의 값의 변화를 관찰한다.

**[한글버전]**

01. Geogebra를 실행시킨다.

02. 입력창에

    **주사위=1**

    입력하여 주사위=1 이라는 수를 만든다.

03. 입력창에

    **주사위던지기=버튼()**

    입력하여 주사위던지기 버튼을 만든다.

04. 주사위던지기 버튼에서 오른쪽 마우스 클릭 --> 설정사항 --> 스크립트 --> 클릭할 때

    1번째 줄 : **값설정[주사위,랜덤[1,6]]**

    을 스크립트 창에 입력 --> 확인 --> 설정사항 창 닫기

05. 이동도구를 선택 --> 주사위던지기 버튼 선택

    *이동도구는 Geogebra 상단 도구들 중 화살표 표시입니다.

06. 주사위던지기 버튼을 선택할 때 마다 주사위 값이 자동으로 변하는 것을 볼 수 있다.

    값설정과 랜덤 이라는 Geogebra 스크립트 명령의 의미를 설명해 준다. 경우에 따라서는 주사위가 수가 기본값이 최댓값이 5인 슬라이더로 되어 6이 표현 안되는 경우가 있다. 이러한 경우 최댓값을 6으로 조정해주면 된다.

07. 입력창에

   **주사위결과모음={}**

   입력하여 주사위결과모음 이라는 빈 리스트를 만든다.

08. 주사위던지기 버튼에서 오른쪽 마우스 클릭 --> 설정사항 --> 스크립트 --> 클릭할 때

    1번째 줄 끝을 선택하고 [Enter]를 입력하면 2번째 줄이 생긴다.

    2번째 줄 : **값설정[주사위결과모음,추가[주사위결과모음,주사위]]**

    을 스크립트 창에 입력 --> 확인 --> 설정사항 창 닫기

09. 이동도구를 선택 --> 주사위던지기 버튼 선택

    *이동도구는 Geogebra 상단 도구들 중 화살표 표시입니다. 

10. 주사위던지기 버튼을 선택할 때 마다 주사위 값이 자동으로 변하는 것을 볼 수 있다. 또한, 주사위결과모음 리스트에 차곡차곡 주사위에서 나온 값들이 쌓이는 것을 볼 수 있다.

11. 입력창에

    **주사위1의개수=세기조건[x==1,주사위결과모음]**

    을 입력하여 주사위 1의 눈의 개수를 자동으로 셀 수 있게 한다.

12. 마찬가지로 입력창에

    **주사위2의개수=세기조건[x==2,주사위결과모음]**

    **주사위3의개수=세기조건[x==3,주사위결과모음]**

    **주사위4의개수=세기조건[x==4,주사위결과모음]**

    **주사위5의개수=세기조건[x==5,주사위결과모음]**

    **주사위6의개수=세기조건[x==6,주사위결과모음]**

    차례로 입력하여 주사위 눈의 개수를 자동으로 셀 수 있게 한다.

13. 마찬가지로 입력창에

    **주사위전체개수=세기조건[x==x,주사위결과모음]**

    차례로 입력하여 주사위 전체개수 자동으로 셀 수 있게 한다.

14. 쓰인 명령어 세기조건에 대하여 이야기 해준다.

    조건으로 쓴 x==1, x==2, ..., x==6에 대하여 알려주고 x==x가 전체개수를 쓰는 조건임을 알려준다.

15. 입력창에

    **주사위1의상대도수=주사위1의개수/주사위전체개수**

    을 입력하여 주사위 1의 눈의 상대도수를 자동으로 계산하게 한다.

16. 마찬가지로 입력창에

    **주사위2의상대도수=주사위2의개수/주사위전체개수**

    **주사위3의상대도수=주사위3의개수/주사위전체개수**

    **주사위4의상대도수=주사위4의개수/주사위전체개수**

    **주사위5의상대도수=주사위5의개수/주사위전체개수**

    **주사위6의상대도수=주사위6의개수/주사위전체개수**

    차례로 입력하여 주사위 눈의 상대도수를 자동으로 계산하게 한다.

17. 각 눈의 상대도수가 분수형태로 되어 있다.

    분수 형태로는 어떤 값에 가까지 감을 보여주기 힘들어 보인다.
