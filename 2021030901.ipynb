    이때 왼쪽 마우스를 누른상태로 Algebra을 넓혀서 ClassBoundary의 내용이

    다 보이도록 창의 너비를 조절한다.

    추후에 적당히 조절하여 Algebra의 내용이 입력내용이 적당히 보이도록 한다.

　

05. Input에

    **NumberOfClass=Length[ClassBoundary]-1**

    입력하여 NumberOfClass라는 수를 만든다.

　

06. Input에

    **ClassOfData=Sequence[{Element[ClassBoundary,t],Element[ClassBoundary,t+1]},t,1,NumberOfClass]**

    입력하여 ClassOfData이라는 list를 만든다.

　

07. Input에

    **FrequencyOfClass={2,4,9,2,3}**

    입력하여 FrequencyOfClass이라는 list를 만든다.

　

08. Input에

    **TheRankValueOfClass=Sequence[mean[Element[ClassOfData,t]],t,1,NumberOfClass]**

    입력하여 TheRankValueOfClass이라는 list를 만든다.

　

09. Input에

    **ApproximateDataConsideringTheRankValue=Flatten[Sequence[Sequence[Element[TheRankValueOfClass,v],t,1,Element[FrequencyOfClass,v]],v,1,NumberOfClass]]**

    입력하여 ApproximateDataConsideringTheRankValue이라는 list를 만든다.

　

10. Input에

    **FrequencyDistributionTable=FrequencyTable[ClassBoundary, ApproximateDataConsideringTheRankValue]**

    입력하여 FrequencyDistributionTable라는 텍스트를 만든다.

　

11. Input에

    **SumOfFrequencies=Sum[FrequencyOfClass]**

    입력하여 SumOfFrequencies이라는 수를 만든다.

　

12. Input에

    **TheRankValueTimesFrequency=Sequence[Element[TheRankValueOfClass,t]*Element[FrequencyOfClass,t],t,1,NumberOfClass]**

    입력하여 TheRankValueTimesFrequency라는 TheRankValueOfClass과 도수를 곱한 list를 만든다.

　

13. Input에

    **SumOfData=Sum[TheRankValueTimesFrequency]**

    입력하여 SumOfData이라는 수를 만든다.

　

14. Input에

    **MeanOfData=SumOfData/Sum[FrequencyOfClass]**

    입력하여 MeanOfData이라는 수를 만든다.

　

15. Input에

    **DeviationOfData=Sequence[Element[TheRankValueOfClass,t]-MeanOfData,t,1,NumberOfClass]**

    입력하여 DeviationOfData이라는 list를 만든다.

　

16. Input에

    **DeviationTimesFrequency=Sequence[Element[DeviationOfData,t]*Element[FrequencyOfClass,t],t,1,NumberOfClass]**

    입력하여 DeviationTimesFrequency이라는 list만든다.

　

17. Input에

    **SumOfDeviationTimesFrequency=Sum[DeviationTimesFrequency]**

    입력하여 SumOfDeviationTimesFrequency이 Sum는 DeviationTimesFrequency다.

　

18. SumOfDeviationTimesFrequency의 "0"이 되는 이유를 생각해본다.

　

19. Input에

   **DeviationSquared=Sequence[Element[DeviationOfData,t]^2,t,1,NumberOfClass]**

    입력하여 DeviationSquared이라는 list를 만든다.

　

20. Input에

    **DeviationSquaredTimesFrequency=Sequence(Element(DeviationSquared, t) * Element(FrequencyOfClass, t), t, 1, NumberOfClass)**

    입력하여 DeviationSquaredTimesFrequency라는 list를 만든다.

　

21. Input에

    **SumOfDeviationSquaredTimesFrequency=Sum[DeviationSquaredTimesFrequency]**

    입력하여 SumOfDeviationSquaredTimesFrequency이라는 수를 만든다.

　

22. Input에

    **DispersionOfData=SumOfDeviationSquaredTimesFrequency/SumOfFrequencies**

    입력하여 DispersionOfData이라는 수를 만든다.

　

23. Input에

    **StandardDeviationOfData=sqrt[DispersionOfData]**

    입력하여 StandardDeviationOfData이라는 수를 만든다.

03. 입력창에

    **자료계급경계={1,3,5,7,9,11}**

    입력하여 자료계급경계라는 리스트를 만든다.

　

　

04. 대수창과 기하창 경계에 마우스를 위치시켰을때 마우스 커서가 좌우화살표로 변한다.

    이때 왼쪽 마우스를 누른상태로 대수창을 넓혀서 자료계급경계의 내용이

    다 보이도록 창의 너비를 조절한다.

    추후에 적당히 조절하여 대수창의 내용이 입력내용이 적당히 보이도록 한다.

　

05. 입력창에

    **계급개수=길이[자료계급경계]-1**

    입력하여 계급개수라는 수를 만든다.

　

06. 입력창에

    **자료계급=수열[{원소[자료계급경계,t],원소[자료계급경계,t+1]},t,1,계급개수]**

    입력하여 자료계급이라는 리스트를 만든다.

　

07. 입력창에

    **자료도수={2,4,9,2,3}**

    입력하여 자료도수이라는 리스트를 만든다.

　

08. 입력창에

    **계급값=수열[평균[원소[자료계급,t]],t,1,계급개수]**

    입력하여 계급값이라는 리스트를 만든다.

　

09. 입력창에

    **계급값원자료=리스트단일화[수열[수열[원소[계급값,v],t,1,원소[자료도수,v]],v,1,계급개수]]**

    입력하여 계급값원자료이라는 리스트를 만든다.

　

10. 입력창에

    **자료도수분포표=도수분포표[자료계급경계, 계급값원자료]**

    입력하여 자료도수분포표라는 텍스트를 만든다.

　

11. 입력창에

    **도수합=합[자료도수]**

    입력하여 도수합이라는 수를 만든다.

　

12. 입력창에

    **계급값곱도수=수열[원소[계급값,t]*원소[자료도수,t],t,1,계급개수]**

    입력하여 계급값곱도수라는 계급값과 도수를 곱한 리스트를 만든다.

　

13. 입력창에

    **자료합=합[계급값곱도수]**

    입력하여 자료합이라는 수를 만든다.

　

14. 입력창에

    **자료평균=자료합/합[자료도수]**

    입력하여 자료평균이라는 수를 만든다.

　

15. 입력창에

    **자료편차=수열[원소[계급값,t]-자료평균,t,1,계급개수]**

    입력하여 자료편차이라는 리스트를 만든다.

　

16. 입력창에

    **편차곱도수=수열[원소[자료편차,t]*원소[자료도수,t],t,1,계급개수]**

    입력하여 자료편차이라는 리스트를 만든다.

　

17. 입력창에

    **편차합=합[편차곱도수]**

    입력하여 편차합이라는 수를 만든다.

　

18. 편차합이 "0"이 되는 이유를 생각해본다.

　

19. 입력창에

    **편차제곱=수열[원소[자료편차,t]^2,t,1,계급개수]**

    입력하여 편차제곱이라는 리스트를 만든다.

　

20. 입력창에

    **편차제곱곱도수=수열[원소[편차제곱,t]*원소[자료도수,t],t,1,계급개수]**

    입력하여 편차제곱곱도수라는 리스트를 만든다.

　

21. 입력창에

    **편차제곱합=합[편차제곱곱도수]**

    입력하여 편차제곱합이라는 수를 만든다.

　

22. 입력창에

    **자료분산=편차제곱합/도수합**

    입력하여 자료분산이라는 수를 만든다.

　

23. 입력창에

    **자료표준편차=sqrt[자료분산]**
